```python
import time
import pandas as pd

df = pd.DataFrame(columns=["n", "runtime"])

def fact(n):
    if n in [0, 1]: return 1
    return n * fact(n-1)

for m in [i for i in range(0, 1000, 200)]:
    total = 0
    for i in range(10):
        start_time = time.perf_counter()
        fact(m)
        end_time = time.perf_counter()
        total += end_time-start_time
    res = (total/10)*1000
    df.loc[len(df)] = [m, res]
```

```python
import matplotlib.pyplot as plt
plt.scatter(x = df['n'], y = df['runtime'])
```
